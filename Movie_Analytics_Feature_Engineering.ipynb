#**Machine Learning Workflow**

```python
new_df = pd.read_csv('movies_metadata.csv.zip')
new_df
```

```python
new_df.select_dtypes(include = "number")
```

```python
def select_columns(df : pd.DataFrame, features : list, target : list = "vote_count" ):

  df = df.dropna().reset_index(drop = True) #handling missing values
  #filtering
  x = df[features]
  y = (df[target] >= 3).astype("int")

  return x,y
```

```python
from sklearn.model_selection import train_test_split
```

```python
features = ["runtime", "budget", "vote_average"]
X, y = select_columns(new_df, features = features)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
X_test
```

```python
#scalling

from sklearn.preprocessing import MinMaxScaler
```

```python
minmax = MinMaxScaler()
```

```python
# Convert 'budget' column to numeric, handling errors
X_train['budget'] = pd.to_numeric(X_train['budget'], errors='coerce')

# Replace non-numeric values with 0 (or another suitable value)
X_train['budget'] = X_train['budget'].fillna(0)

# Now fit the MinMaxScaler
minmax.fit(X_train)
```

```python
# Select only numeric features for scaling
numeric_features = ['runtime', 'budget', 'vote_average']
X_train_numeric = X_train[numeric_features]
X_test_numeric = X_test[numeric_features]

# Convert 'budget' column to numeric in both training and testing sets
X_train_numeric['budget'] = pd.to_numeric(X_train_numeric['budget'], errors='coerce').fillna(0)
X_test_numeric['budget'] = pd.to_numeric(X_test_numeric['budget'], errors='coerce').fillna(0)

# Fit MinMaxScaler on training data and transform both training and testing data
minmax = MinMaxScaler()
X_train_scaled = minmax.fit_transform(X_train_numeric)
X_test_scaled = minmax.transform(X_test_numeric)
```

```python
def scaling(df, scaler):

  df = df.dropna().reset_index(drop = True) #handling missing values
  numeric_features = ['runtime', 'budget', 'vote_average']  # Replace with your actual numeric feature names
  df_numeric = df[numeric_features]

  # Convert numeric features to numeric type
  df_numeric['budget'] = pd.to_numeric(df_numeric['budget'], errors='coerce').fillna(0)

  # Transform the data using the scaler
  scaled_array = scaler.transform(df_numeric)

  # Create a new DataFrame with scaled values and original column names
  scaled_df = pd.DataFrame(scaled_array, columns=numeric_features, index=df.index)

  # Concatenate the scaled numeric features with the original DataFrame
  final_df = pd.concat([df.drop(numeric_features, axis=1), scaled_df], axis=1)

  return final_df
```

```python
X_train_scaled = scaling(X_train, minmax)
X_test_scaled = scaling(X_test, minmax)
```
